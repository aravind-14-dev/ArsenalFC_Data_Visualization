```python
#Importing the required libraries and loading the stats.csv file
import pandas as pd
import numpy as np
data = pd.read_csv(r"stats.csv")
```

```python
#Viewing the Top 5 rows of the dataset
data.head()
```

```python
#Extracting the data that contains only Arsenal's stats
data_arsenal = data[data.team=='Arsenal']
data_arsenal
```

```python
#Feature building - extracting only the goals and season column from the dataset
feature_goals = ['goals' , 'season']   
data_goalsEseason = data_arsenal[feature_goals]  #data that provides number of goals scored every season
```

```python
#Feature building - extracting only the goals conceded and season column from the dataset
feature_conceded = ['goals_conceded','season']   
data_concededEseason = data_arsenal[feature_conceded]   #data that provides number of goals scored every season
```

```python
#Plotting the features using matplot library
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12,8]
plt.plot(data_goalsEseason.season, data_goalsEseason.goals, label = 'Scored', marker = 'o')
plt.plot(data_concededEseason.season, data_concededEseason.goals_conceded, label = 'Conceded', marker = 'o')
plt.ylim(0,150)
plt.grid(which='major', axis = 'both', linestyle = '-.', linewidth = 0.75)
plt.xticks(rotation=60, fontsize = 14)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 14)
plt.title('Scored and Conceded Vs Season', fontsize = 20)

```

```python
#Importing results.csv file
data1 = pd.read_csv("results.csv")
data1.head()
```

```python
#Choosing Big 5 clubs as home team to visualize away performances against them
home_Team1 = data1[data1['home_team'].isin(["Manchester United","Chelsea", "Liverpool", "Manchester City", "Tottenham Hotspur"])]
away_Team = home_Team1[home_Team1.away_team == "Arsenal"]  #Extracting the Arsenal data as away team

feature_opponentG = ['home_goals','season'] #Feature to extract the no of goals scored by Big 5
opp_goalsVseason = away_Team[feature_opponentG]  #Extracting the data that tells the no of goals scored by Big 5


feature_awayG = ['away_goals','season'] #Feature to extract the no of goals scored by Arsenal against Big 5
away_goalsVseason = away_Team[feature_awayG] #Extracting the data that tells the no of goals scored by Arsenal

print(opp_goalsVseason.head())
print(away_goalsVseason.head())

#You can see that data is not aggregated

```

```python
goals_conceded = opp_goalsVseason.groupby('season')['home_goals'].sum().reset_index(name='home_goals') #Aggregates the data for each season
goals_scored = away_goalsVseason.groupby('season')['away_goals'].sum().reset_index(name='away_goals')
print(goals_conceded)
print(goals_scored)
```

```python
#Visualizing the features that we built
plt.rcParams['figure.figsize'] = [12, 8]
plt.plot(goals_scored.season, goals_scored.away_goals, label = 'GS', marker='o')
plt.plot(goals_conceded.season, goals_conceded.home_goals, label = 'GC', marker='o')
plt.ylim(0,30)
plt.grid(which='major', axis='both', linestyle='-.', linewidth=0.75)
plt.xticks(rotation=60, fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.title('Goals Scored and Conceded against top 5 clubs in away games v/s Season', fontsize = 20)
```

```python
#To visualize Goals conceded against each of the Big 5 clubs in away games v/s Season
#Extracting the relevant data again for new visualization
home_Team1 = data1[data1['home_team'].isin(["Manchester United","Chelsea", "Liverpool", "Manchester City", "Tottenham Hotspur"])]
away_Team = home_Team1[home_Team1.away_team == "Arsenal"]

fig, ax = plt.subplots(figsize=(10,8)) 
teams = away_Team['home_team'].drop_duplicates()  #Removing duplicate rows if any
colors = ["Blue","Red","Green","Black","Orange"]  #Defining colors to represent each opposition in list
margin_bottom = np.zeros(len(away_Team['season'].drop_duplicates())) #counter to continuously plot 5 teams in a single bar graph

for num,team in enumerate(teams):  #loop into each team and plot their data
    values = list(away_Team[away_Team['home_team']==team].loc[:,'home_goals'])
    away_Team[away_Team['home_team']==team].plot.bar(x='season',y='home_goals', ax= ax, stacked = True,bottom = margin_bottom, color = colors[num], label=team)
    margin_bottom += values

plt.title('Goals conceded against each of the Big 5 clubs in away games v/s Season', fontsize = 20)
plt.show()
#print(away_Team)



```
